<h3>Ruina del jugador</h3>

```python
def moneda():
    x = random()
    if x <= 0.5:
        return -1
    else:
        return 1
```

La lista contiene el dinero del jugador

```python
def ruina(N):
    L = [N]
    while N > 0:
        N += moneda()
        L.append(N)
    return L
```

<h3>Jugador precavido</h3>
<p>AHORA fijamos un tope superior, es decir el jugador abandona cuando llega a un cierto valor de su 'riqueza':</p>

```python
def ruina_s(N,M):
    #L = [N]
    while M > N > 0:
        N += moneda()
    #   L.append(N)
    return N,M
```

```python
ruina_s(100,1100)
```

```python
def prob_no_ruina(N,M,R):
    cont = 0
    for int in srange(R):
            ruina,exito=ruina_s(N,M)
            if ruina != 0:
                cont += 1
    return (cont/R).n()
'''Dividimos los casos en que no se arruina entre los totales'''
```

```python
prob_no_ruina(100,110,10000)
```

```python
LL=[(100+k*10,prob_no_ruina(100,100+k*10,1000)) for k in srange(10)]
```

<h3>Casino limitado</h3>

Introducimos un l√≠mite de dinero al casino.

```python
def casino_limitado(RC,RJ):
    while (RC>0 and RJ>0):
        x = moneda()
        if x == 1:
            RJ += 1
            RC -= 1
        else:
            RJ -= 1
            RC += 1
    return RC,RJ
```

```python
casino_limitado(200,100)
```

```python
def prob_ruina_jugador(RCI,RJI,R):
    cont = 0
    for muda in xrange(R):
        rcasino,rjugador = casino_limitado(RCI,RJI)
        if rjugador == 0:
            cont += 1
    return (cont/R).n()
```

```python
prob_ruina_jugador(100,100,1000)
```

```python
prob_ruina_jugador(200,100,1000)
```
