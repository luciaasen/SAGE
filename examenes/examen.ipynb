```python
# ## Ej 1
# 
# La funcion criba que nos dan devuelve una lista de primos. Como me piden decidir si n es primo, creamos una funcion que verifique si n esta en la lista.
# 
# Ademas, creamos funcion wilson
# 

# In[28]:

def criba(n):
        aux = [True]*int(n)
        aux[0] = False
        aux[1] = False
        for i in xrange(2,floor(sqrt(n))+1):
            if aux[i]:
                for j in xrange(i*i,n,i):
                    aux[j] = False
        return [k for k in xrange(n) if aux[k]==True]

def es_primo(n):
    if n in criba(n+1):
        return True
    return False

    
def wilson(n):
    Zn = Integers(n)
    n1 = Zn(factorial(n-1))
    n2 = Zn(-1)
    if n1 == n2:
        return True
    return False


def lista_rand(n1, n2, tam):
    L = []
    for i in srange(tam):
        L.append(randint(n1, n2))
    L.sort()
    return L

def twilson(n):
    t = cputime(wilson(n))
    return t

def tcriba(n):
    t = cputime(es_primo(n))
    return t

def guarda_tiempos(n1, n2, tam):
    
    L = lista_rand(n1, n2, tam)
    
    Lcriba = []
    for item in L:
        Lcriba.append((item, tcriba(item)))
    
    Lwilson = []
    for item in L:
        Lwilson.append((item, twilson(item)))
    
    return Lcriba, Lwilson

Lcriba, Lwilson = guarda_tiempos(10**4, 10**5, 100)

Lfinal = []
Lfinal.append(line2d(Lcriba))
Lfinal.append(line2d(Lwilson))
show(sum(Lfinal))
show(Lcriba[0])


# Al tener el valor de Lcriba[0], sabemos que la grÃ¡fica de Lcriba, al estar por debajo, es la que tiene en todo momento tiempos mas bajos para un rango de [N1, N2] = [10^4, 10^5]
# 
# Por esto concluimos que el mÃ©todo de la criba es, en este caso, mÃ¡s eficiente.
# 
# Probemos con otro [N1, N2]: 

# In[29]:

Lcriba2, Lwilson2 = guarda_tiempos(10**5, 10**6, 100)

Lfinal2 = []
Lfinal2.append(line2d(Lcriba2))
Lfinal2.append(line2d(Lwilson2))
show(sum(Lfinal2))
show(Lcriba2[0])


# Los resultados no cambian para este nuevo rango:  
# El metodo de la criba sigue siendo mÃ¡s eficiente que el de wilson, y con este nuevo rango el tiempo de wilson llega casi a duplicar el de la criba.
# 
# Probamos con N1, N2 menores:

# In[31]:

Lcriba3, Lwilson3 = guarda_tiempos(10**2, 10**3, 100)

Lfinal3 = []
Lfinal3.append(line2d(Lcriba3))
Lfinal3.append(line2d(Lwilson3))
show(sum(Lfinal3))
show(Lcriba3[0])


# Aunque la diferencia entre ambos metodos para numeros mas pequeÃ±os es casi despreciable, la criba sigue siendo mas eficiente.
# 
#  
# Hacemos una ultima prueba, con numeros mayores, para asegurar que wilson sigue siendo el peor metodo:
# 
# 

# In[33]:

Lcriba4, Lwilson4 = guarda_tiempos(10*6, 10**7, 20)

Lfinal4 = []
Lfinal4.append(line2d(Lcriba4))
Lfinal4.append(line2d(Lwilson4))
show(sum(Lfinal4))
show(Lcriba4[0])


# El metodo de la criba sigue siendo Ã³ptimo.
# 
# Probamos un ultimo caso, con valores de N1 y N2 minimos:

# In[36]:

Lcriba5, Lwilson5 = guarda_tiempos(10*1, 10**2, 50)

Lfinal5 = []
Lfinal5.append(line2d(Lcriba5))
Lfinal5.append(line2d(Lwilson5))
show(sum(Lfinal5))
show(Lcriba5[0])
show(Lwilson5[0])


# Con esta ultima grafica llegamos a la conlcusion de que el metodo de la criba es definitivamente mejor que el de wilson, y que esta diferencia aumenta conforme aumenta nuestro rango [N1, N2] donde buscamos primos. (Para este ultimo ejemplo la diferencia es minima)
# 
# Por tanto, no puedo discutir el posible cambio que indica el ejercicio (que en diferentes rangos, el metodo mas eficiente varie)

# In[ ]:




# In[ ]:
```
